### Задание 3

Добавьте в функцию расчет среднего значения словаря (разделите сумму на количество значений). Данная цифра также должна быть в return функции.

Добавьте проверку на наполненность словаря с помощью конструкции try-except: функция должна принтить "Словарь должен быть не пустой!" в случае если ей был передан пустой словарь в качестве аргумента.

```python
def func_calc(dct):
  total=0
# Считаем сумму с обработкой не-чисел
  for i in dct.values():
    try:
        total=total+i
    except TypeError:
       print(f"Пропускаем не-цифровое значение: {i}")

# Считаем среднее с проверкой на пустой словарь
  try:
      avg = total/len(dct)
  except ZeroDivisionError:
        print("Словарь должен быть не пустой!")
        return None, None

  return total, avg


```

```python
func_calc(dct_2)
```

### Задание 4

Добавьте в функцию еще один аргумент: пороговое значение thr (от англ. Threshold).
    
В расчете суммы и среднего должны участвовать только те значения словаря, которые превышают (строго) thr. Задайте параметру thr значение по умолчанию: 0.

Обратите внимание на последовательность вложенности: for -> try -> if

```python
def func_calc(dct, thr = 0):
  total=0
  count=0                  # Счётчик элементов, подходящих под условие

  for i in dct.values():
    try:                   # Сначала проверяем, является ли значение числом и больше ли оно thr
        if i > thr:
                total += i
                count += 1
    except TypeError:
            print(f"Пропускаем не-цифровое значение: {i}")

  try:                     # Считаем среднее, используя новый счетчик count
        avg = total / count
  except ZeroDivisionError:
        print("Нет значений, подходящих под условие или словарь пуст!")
        return None, None

  return total, avg
```

Протестируйте вашу функцию на словаре dct_4

```python
dct_4 = {'a':1, 'b':[1,2,3], 'c':7, 'd':-8, 'e':'f'}
```

```python
func_calc(dct_4)
```
