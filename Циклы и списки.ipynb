### Задача 3
Используя список payments из предыдущей задачи, найдите средний платёж.

**Задание:**

Рассчитайте средний платёж двумя способами:

Вручную — с помощью цикла:

+ посчитайте сумму всех платежей

+ посчитайте количество платежей

+ найдите среднее как отношение суммы к количеству
(не используйте функции sum() и len())

С помощью встроенных функций — используя sum() и len()

Сравните результаты: они должны совпадать.

```python
# Решение
payments = [1200, 3000, 120000, 10800, 3000, 18300]
```

```python
total_sum = 0
for payment in payments:
    total_sum += payment
print("Общая сумма(через цикл):", total_sum)

cnt_payments = 0
for payment in payments:
    cnt_payments +=1
print("Количество платежей(через цикл):", cnt_payments)

#Найдем средний платеж как отношение суммы к количеству
avg_pay = total_sum/cnt_payments
print("Средняя сумма платежа(через цикл):", avg_pay)

#Рассчитаем средний платеж через встроенные функции
sum_payments = sum(payments)
print("Общая сумма(через sum):", sum_payments)

cnt_payments = len(payments)
print("Общая сумма(через len):", cnt_payments)

avg_payments = sum_payments/cnt_payments
print("Средняя сумма платежа:", avg_payments)
```

### Задача 4.

Менеджер хочет понять, какие платежи были крупными.

**Задание:**

Создайте новый список *big_payments*.

Добавьте в него только те значения из списка payments, которые больше 10000 рублей.

Выведите список *big_payments*.

Затем выведите индексы этих платежей в исходном списке payments вместе с их значениями в формате:
индекс: значение

**Подсказка:**

+ используйте цикл for

+ для получения индекса используйте range(len(payments))

```python
# Решение
payments = [1200, 3000, 120000, 10800, 3000, 18300]
```

```python
big_payments = []
for payment in payments:
    if payment >= 10000:
      big_payments.append(payment)
print("Список платежей, которые больше 10000 рублей:", big_payments)

print("Индексы и значения платежей:")
for i in range(len(payments)):
    if payments[i] > 10000:
      print(i, ":", payments[i])

```

Каждое значение выведите на отдельной строке.

**Подсказка:**

+ используйте цикл for, чтобы перебрать элементы списка

+ если элемент — число, выведите его

+ если элемент — список или кортеж, пройдитесь по нему с помощью вложенного цикла

+ игнорируйте все нечисловые значения

```python
clients_payments = [
    "Иван",
    1000,
    [1200, 300, 800],
    "Мария",
    "ошибка ввода",
    (4000, 5000)
]
```

```python
# Решение
elements_list = []
for i in clients_payments:
    if (type(i)==list) | (type(i)==tuple):
        for j in i:
            elements_list.append(j)
    else:
        elements_list.append(i)

print(elements_list)

for i in elements_list:
    if type(i) == int:
        print(i)
```
