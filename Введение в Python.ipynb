### Задание 2. Анализ стихотворения

Вам дана строка second_string

```python
second_string = """Мчатся тучи, вьются тучи;
Невидимкою луна
Освещает снег летучий;
Мутно небо, ночь мутна.
Еду, еду в чистом поле;
Колокольчик дин-дин-дин.
Страшно, страшно поневоле
Средь неведомых равнин!"""
```

####   Задание 2.1

Переход на следующую строку является непечатным символом "\n".\
Рассчитайте количество строк в тексте как количество переходов между строками плюс один.

Подсказка: передайте символ "\n" в качестве аргумента метода count.

```python
cnt_line = second_string.count("\n")+1
print(cnt_line)
```

####   Задание 2.2

 1. Подсчитайте количество точек, запятых, точек с запятой, восклицательных знаков и тире.
 2. Количество для каждого знака препинания запишите в отдельную переменную.
 3. Рассчитайте количество пробелов и запишите в отдельную переменную.
 4. Определите "буквенную длину строки" с помощью вычитания количеств знаков препинания, пробелов и нечитаемых символов из общей длины
 5. Определите среднюю длину строки в буквах (не считая знаков препинания).

```python
a = second_string.count(".")
b = second_string.count(",")
c = second_string.count(";")
d = second_string.count("!")
e = second_string.count("-")
f = second_string.count(" ")
g = second_string.count("\n")+1
h = len(second_string)

string_length = h - (a + b + c + d + e + f + g)
m=string_length/g

print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
print(h)
print(string_length)
print(m)
```
