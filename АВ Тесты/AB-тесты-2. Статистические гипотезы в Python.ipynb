```python
path = '/content/drive/MyDrive/животные.xlsx'
an = pd.read_excel(path)
an.head()
```

```python
#Постройте группировку, по которой видно, какое количество заказов было для каждой пары «животное — порода» и какой средний чек был по этой паре
an_gr = an.groupby(['animal','species']).agg(cnt_order = ('check', 'count'), avg_check = ('check', 'mean')).\
reset_index().sort_values('animal')
an_gr
```

```python
#Найдите заказы с аномально высокими чеками, например которые превышают среднее значение на 1,5 стандартных отклонения.
#Находим среднее значение и стандартное отклонение
mean_check = an['check'].mean()
std_check = an['check'].std()
print(mean_check)
print(std_check)
```

```python
# Считаем границу: среднее + 1.5 * стандартное отклонение
limit = mean_check + 1.5 * std_check
limit
```

```python
#Определяем аномально высокие чеки
anomal = an[an['check'] > limit]
```

```python
#Выводим результаты
print(f"Граница аномальных чеков: {limit}")
print(f"Найдено аномальных заказов: {len(anomal)}")
```

```python
#Найдите медианный чек услуги для каждого животного в каждом возрасте.
an_gr_median = an.groupby(['animal','age']).agg(median_check = ('check','median')).reset_index()
an_gr_median
```

Используя функцию ttest_ind, проверьте равенство средних двух выборок и решите, отвергаете или принимаете нулевую гипотезу.

1.Какой средний чек выше: по собакам или по кошкам?

2.Какой средний чек выше: по экзотическим животным или по собакам?

3.Какой средний чек выше: по экзотическим животным или по кошкам?

4.Существуют ли сегменты (разбиение по породе), в рамках которых средний чек по кошкам больше, чем по собакам?

```python
#Создаем новые таблицы отдельно
dog = an[an['animal']=='Собака']
cat = an[an['animal']=='Кошка']
exotic = an[an['animal']=='Экзотическое животное']
```

```python
#1.Какой средний чек выше: по собакам или по кошкам?
#Находим среднее значение
dog_mean = dog['check'].mean()
cat_mean = cat['check'].mean()
exotic_mean = exotic['check'].mean()
print(f'Среднее значение чека по собакам:{dog_mean}')
print(f'Среднее значение чека по кошкам:{cat_mean}')
print(f'Среднее значение чека по экзотическим животным:{exotic_mean}')
```

```python
#Длинна данных выборки
print(len(dog))
print(len(cat))
print(len(exotic))
```

```python
#Решаем с помощью функции ttest_ind и принимаем решение
s, p = stats.ttest_ind(dog['check'], cat['check'])
print(s)
print(p)
```

```python
s, p = stats.ttest_ind(exotic['check'], dog['check'])
print(s)
print(p)
```

```python
s, p = stats.ttest_ind(exotic['check'], cat['check'])
print(s)
print(p)
```

Оталось еще два вопроса.

4.Существуют ли сегменты (разбиение по породе), в рамках которых средний чек по кошкам больше, чем по собакам?

5.Какой средний чек выше: по экзотическим животным или по домашним породистым кошкам?

```python
#4.Существуют ли сегменты (разбиение по породе), в рамках которых средний чек по кошкам больше, чем по собакам?
#Создаем таблицы
cat_home = an[(an['animal'] == 'Кошка') & (an['species'] == 'Домашняя обыкновенная')]
cat_ped = an[(an['animal'] == 'Кошка') & (an['species'] == 'Домашняя породистая')]
cat_str = an[(an['animal'] == 'Кошка') & (an['species'] == 'Уличная')]

dog_home = an[(an['animal'] == 'Собака') & (an['species'] == 'Домашняя обыкновенная')]
dog_ped = an[(an['animal'] == 'Собака') & (an['species'] == 'Домашняя породистая')]
dog_str = an[(an['animal'] == 'Собака') & (an['species'] == 'Уличная')]
```

```python
#Вводим функцию, которая проверит гипотезу:
def mytest(x1, x2, pvalue=0.05):
    print('разница между средними значениями:', x1.mean()-x2.mean())
    s, p = stats.ttest_ind(x1, x2)
    if p<=pvalue:
        print('Гипотеза отвергается, разница стат. значима')
    else:
        print('Гипотеза не отвергается, средние равны')
```

```python
#Запускаем функцию на проверку кошки и собаки порода домашняя обыкновенная
mytest(cat_home['check'], dog_home['check'])
```

```python
#Запускаем функцию на проверку кошки и собаки порода домашняя породистая
mytest(cat_ped['check'], dog_ped['check'])
```

```python
cat_m = cat_ped['check'].mean()
dog_m = dog_ped['check'].mean()
print(cat_m)
print(dog_m)
```

```python
#Запускаем функцию на проверку кошки и собаки порода уличная
mytest(cat_str['check'], dog_str['check'])
```

Ответ на вопрос Существуют ли сегменты (разбиение по породе), в рамках которых средний чек по кошкам больше, чем по собакам?

У домашней породистой кошки средний чек больше, чем у домашней породистой собаки.

```python
#5.Какой средний чек выше: по экзотическим животным или по домашним породистым кошкам?
cat_m = cat_ped['check'].mean() #средний чек домашней породистой кошки
exotic_m = exotic['check'].mean() #средний чек экзотических животных
print(cat_m)
print(exotic_m)
```

```python
s, p = stats.ttest_ind(exotic['check'], cat_ped['check'])
print(s)
print(p)
```

```python
mytest(cat_ped['check'], exotic['check'])
```

Статистически значимой разницы между средними чеками экзотических животных и домашних породистых кошек не обнаружено (p > 0.05). Их средние чеки можно считать равными
