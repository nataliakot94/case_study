### Задание 3

Воспроизведите результат следующего SQL-кода

~~~ sql
select   workout_id
       , trainer_id
from workouts
where  workout_schedule_type in ('trial', 'regular')
  and trainer_department <> 'Sales'
order by trainer_id
limit 5
~~~

```python
#####
workouts[(workouts['workout_schedule_type'].isin(['trial', 'regular'])) & (workouts['trainer_department'] != 'Sales')][['workout_id','trainer_id']].sort_values('trainer_id').head(5)
```

### Задание 4

Воспроизведите результат следующего SQL-кода

~~~ sql
select   *
from workouts
where  trainer_department like '%Spartacus%'
   and workout_schedule_type not in ('trial', 'single')
order by client_id desc
limit 15
~~~

```python
#####
workouts[
         (workouts['trainer_department'].str.contains('Spartacus', na=False))
        &
         (~workouts['workout_schedule_type'].isin(['trial', 'single']))
        ].sort_values('client_id', ascending=False).head(15)
```

### Задание 5

~~~ sql
select   *
from workouts
where  status = 'success'
    or client_status = 'active' and workout_schedule_type = 'regular'
order by trainer_id
limit 5
~~~

```python
#####
workouts[(workouts['status']=='success') | (workouts['client_status'] == 'active') & (workouts['workout_schedule_type']=='regular')].sort_values('trainer_id').head(5)
```
