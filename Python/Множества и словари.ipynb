# Задание 3. Цикл while

```python
##############################################   Задание 3.1   #################################################################

### С помощью цикла while записывайте в список lst_sum все элементы с конца списка lst_init, пока сумма элементов в lst_sum не достигнет 100
### Вам потребуются методы списков append и pop

lst_init = [4, 17, 22, 10, 8, 15, 8, 9, 2, 32, 13, 11, 12, 10, 9, 12, 22, 23, 13, 11, 10, 7, 42, 43, 14, 15, 16]
lst_sum = []

while sum(lst_sum)<100:
  lst_sum.append(lst_init.pop())
print(lst_sum)
```

```python
##############################################   Задание 3.2   #################################################################

### Создайте переменную bin_var, которая является параметром и может принимать значения 1 или 0.
### Если она равняется другому, то выводится сообщение "Переменная bin_var должна равняться или 1, или 0!"
### Если она равняется нулю, то выполняется действие из задания 3.1
### Если она равняется единице, то действие из задания 3.1 выполняется не с конца списка lst_init, а с начала
### Для этого вам потребуется метод списка reverse(). Не забудьте пересоздать развернутый наоборот список внутри условия

lst_init = [4, 17, 22, 10, 8, 15, 8, 9, 2, 32, 13, 11, 12, 10, 9, 12, 22, 23, 13, 11, 10, 7, 42, 43, 14, 15, 16]
lst_sum = []

bin_var = 0
if bin_var not in [0,1]:
      print('error')
else:
      if bin_var == 1:
        lst_init=lst_init[::-1]
      while sum(lst_sum)<100:
        lst_sum.append(lst_init.pop())
print(lst_sum)

```
