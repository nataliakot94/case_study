```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

```python
# Задача 1. Рассчитайте средний возраст по всему списку age_lst

avg_age = sum(age_lst)/len(age_lst)
print(f"Средний возраст одного пользователя {round(avg_age, 2)} лет.")
```

```python
#Задача 2. Рассчитайте средний возраст для посетителей, которые моложе 50 лет
#Сортируем список по возрастанию.
print(sorted(age_lst))
```

```python
# Считаем средний возраст посетителей, которые моложе 50 лет.
sum_age = 0       # Переменная для накопления суммы возрастов
cnt_age = 0       # Переменная для подсчета количества подходящих людей
for i in age_lst: # Перебираем каждый возраст в списке age_lst
  if i < 50:      # Проверяем, что возраст меньше 50 лет
    sum_age += i  # Прибавляем возраст к общей сумме
    cnt_age += 1  # Увеличиваем счетчик людей на 1
avg_age = sum_age/cnt_age # Вычисляем среднее арифметическое (сумма / количество)
# Выводим результат, округляя средний возраст до двух знаков после запятой
print(f"Средний возраст посетителей, которые моложе 50 лет - {round(avg_age,2)} лет.")
```

```python
# Задача 3. Оберните предыдущий код в функцию, которая на вход принимает список и пороговое значение
def calc_avg_age (lst, thresh):  # Объявляем функцию, которая принимает список возрастов (lst) и порог возраста (thresh)
  sum_age = 0      # Переменная для сложения возрастов
  cnt_age = 0      # Переменная для подсчета количества людей
  for i in lst:    # Проходим циклом по всем элементам переданного списка
    if i < thresh: # Если текущий возраст меньше заданного порога
      sum_age += i # Добавляем его к общей сумме
      cnt_age += 1 # Увеличиваем счетчик найденных людей
  avg_age = sum_age/cnt_age   # Считаем среднее значение (делим сумму на количество)
  print(f"Средний возраст посетителей, которые моложе 50 лет - {round(avg_age,2)} лет.") # Выводим результат, округляя средний возраст до двух знаков после запятой
  return round(avg_age,2) # Возвращаем округленный результат для дальнейшего использования в программе
```

```python
calc_avg_age (age_lst, 50)
```

```python
# Задача 4. Рассчитайте среднее значение для всех элементов списка, за исключением тех, которые лежат вне интервала [Q1-1,5IQR; Q3+1,5IQR]
import numpy as np # Подключаем библиотеку NumPy под псевдонимом np
```

```python
np.mean(age_lst) # Вычисляем среднее арифметическое всех значений в списке age_lst при помощи NumPy
```

```python
np.median(age_lst)  # Вычисляем медиану — значение, которое находится ровно в середине отсортированного списка возрастов
```

```python
q1 = np.percentile(age_lst, 25) # Находим 25-й перцентиль (первый квартиль)
q2 = np.percentile(age_lst, 50) # Находим 50-й перцентиль: это медиана
q3 = np.percentile(age_lst, 75) # Находим 75-й перцентиль (третий квартиль)
p95 = np.percentile(age_lst, 95)# Находим 95-й перцентиль: граница, которую не превышают 95% посетителей
print(q1, q2, q3, p95)          # Выводим все рассчитанные показатели в одну строку
```

```python
iqr = q3-q1               # Вычисляем межквартильный размах (Interquartile Range) — разницу между 75-м и 25-м перцентилями
lower_bound = q1-1.5*iqr  # Рассчитываем нижнюю границу: всё, что меньше этого значения, считается аномально низким (выбросом)
upper_bound = q3+1.5*iqr  # Рассчитываем верхнюю границу: всё, что больше этого значения, считается аномально высоким (выбросом)
print(lower_bound)        # Выводим на экран значение нижней границы
print(upper_bound)        # Выводим на экран значение верхней границы
```

```python
sum_age = 0               # Создаем переменную для накопления суммы возрастов, входящих в диапазон
cnt_age = 0               # Создаем счетчик для подсчета количества людей без учета выбросов
for i in age_lst:         # Запускаем цикл для перебора каждого значения в исходном списке возрастов
   if (i<upper_bound) and (i>lower_bound):  # Проверяем, находится ли текущий возраст внутри рассчитанных границ (не является ли он выбросом)
    sum_age += i          # Если условие верно, прибавляем этот возраст к общей сумме
    cnt_age += 1          # Увеличиваем счетчик подходящих под условие людей на единицу
avg_age = sum_age/cnt_age # Вычисляем среднее арифметическое, деля очищенную сумму на количество людей
print(f"Средний возраст посетителей без аномалий - {round(avg_age,2)} лет.") # Выводим результат
```

```python
#Задача 5 (со звёздочкой). Оберните предыдущий код в функцию, которая принимает на вход список и рассчитывает средний возраст, исключая из расчётов выбросы
def calc_avg_age (lst):
  try:
      q1 = np.percentile(age_lst, 25) # Находим 25-й перцентиль (первый квартиль)
      q3 = np.percentile(age_lst, 75) # Находим 75-й перцентиль (третий квартиль)
      iqr = q3-q1                     # Вычисляем межквартильный размах (Interquartile Range) — разницу между 75-м и 25-м перцентилями
      lower_bound = q1-1.5*iqr        # Рассчитываем нижнюю границу: всё, что меньше этого значения, считается аномально низким (выбросом)
      upper_bound = q3+1.5*iqr        # Рассчитываем верхнюю границу: всё, что больше этого значения, считается аномально высоким (выбросом)
      sum_age = 0               # Создаем переменную для накопления суммы возрастов, входящих в диапазон
      cnt_age = 0               # Создаем счетчик для подсчета количества людей без учета выбросов
      for i in age_lst:         # Запускаем цикл для перебора каждого значения в исходном списке возрастов
        if (i<upper_bound) and (i>lower_bound):  # Проверяем, находится ли текущий возраст внутри рассчитанных границ (не является ли он выбросом)
          sum_age += i          # Если условие верно, прибавляем этот возраст к общей сумме
          cnt_age += 1          # Увеличиваем счетчик подходящих под условие людей на единицу
      avg_age = sum_age/cnt_age # Вычисляем среднее арифметическое, деля очищенную сумму на количество людей
      print(f"Средний возраст посетителей без аномалий - {round(avg_age,2)} лет.") # Выводим результат
  except:
    print('Список не должен содержать нечисловых значений или быть пустым')
```

```python
calc_avg_age(age_lst)
```
