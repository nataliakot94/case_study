```python
##############################################   Задание 1.6   #################################################################

##### Создайте дополнительную ось Y справа (используйте twinx())
##### На этой оси постройте столбчатую диаграмму, отображающую разницу Y-Z
##### Сделайте столбцы полупрозрачными: alpha=0.3
##### Добавьте label='difference' для отображения в легенде
##### Расположите легенды:
#####   - Основную легенду — снизу
#####   - Легенду дополнительной оси — справа-сверху


# 1. Размер таблицы
fig, ax = plt.subplots(figsize=(7, 7))

# 2. Исходные переменные
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [8, 5, 6, 10, 18, 26, 18, 22, 23]
z = [6, 9, 13, 13, 16, 12, 20, 18, 17]

# Считаем разницу для столбчатой диаграммы
difference = [yi - zi for yi, zi in zip(y, z)]

# 3. Рисуем кривые и точки через объект ax
ax.plot(x, y, color='g')
ax.scatter(x, y, color='red', edgecolor='red')
ax.plot(x, z, color='purple')

# 4. Добавляем вертикальные линии через ax.axvline
ax.axvline(x=2, color='red', linestyle='--', label='Минимум')
ax.axvline(x=6, color='red', linestyle='-', label='Максимум')
ax.axvline(x=1, color='black', linestyle='--', label='Минимум Z')
ax.axvline(x=7, color='black', linestyle='-', label='Максимум Z')
# 5. Создаем вторую ось Y справа и строим столбчатую диаграмму
ax2 = ax.twinx()
ax2.bar(x, difference, alpha=0.3, color='gray', label='difference')

# 6. Настраиваем оформление для обеих осей
ax.set_title('Мой первый график')
ax.set_xlabel('Ряд Х')
ax.set_ylabel('Ряд Y')     # Подпись левой оси Y
ax2.set_ylabel('Разница Y-Z')  # Подпись правой оси Y
ax.grid(True)

# 7. Настройка и расположение легенд
ax.legend(loc='lower center')  # Основная легенда — снизу по центру
ax2.legend(loc='upper right')  # Легенда дополнительной оси — справа-сверху

# 8. Итог
plt.show()
```

```python
##############################################   Задание 2.1   #################################################################

##### Импортните csv-файл под названием Purchase_List
##### Приведите date_purchase к формату даты с помощью функции to_datetime
##### Создайте колонку датафрейма с месяцесем оплат (числом, от 1 до 12)
##### Сделайте группировку: количество оплат и сумма оплат для каждого месяца (создайте новый группировочный датафрейм)
##### Постройте визуализацию: для каждого месяца покупки по левой оси постройте сумму оплат, а по правой - количество оплат
##### Не забудьте оформление графика (заглавие, подпись осей, раздельная легенда, сетка)

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import datetime as dt

path = '/content/drive/MyDrive/урок 8/Purchase_List.csv'
purchase_list = pd.read_csv(path)
purchase_list.head()


```

```python
purchase_list['date_purchase'] = pd.to_datetime(purchase_list['date_purchase'])
purchase_list['date_month'] = purchase_list['date_purchase'].dt.month
purchase_list.head()
```

```python
#Сделайте группировку: количество оплат и сумма оплат для каждого месяца (создайте новый группировочный датафрейм)
amt_month = purchase_list.groupby('date_month').agg(
    count_purchases = ('date_purchase', 'count'),
    sum_purchases = ('amt_pay', 'sum')).reset_index()
amt_month
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(amt_month['date_month'], amt_month['sum_purchases'], marker='o', label='Сумма оплат')
ax1.set_ylabel('Сумма оплат в рублях')
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
ax2.bar(amt_month['date_month'], amt_month['count_purchases'], alpha=0.3, color='blue', label='Количество оплат')
ax2.set_ylabel('Количество оплат в штуках')
ax2.tick_params(axis='y')

ax1.set_title('Анализ покупок по месяцам')
ax1.set_xlabel('Номер месяца')
ax1.set_xticks(range(5, 13))
ax1.grid(True, linestyle='--', alpha=0.6)

ax1.legend(loc='lower left')
ax2.legend(loc='upper right')

plt.show()
```

```python
amt_month['avg_purchase'] = amt_month['sum_purchases']/amt_month['count_purchases']

import matplotlib.pyplot as plt

fig, am = plt.subplots(figsize=(10, 6))

am.scatter(amt_month['date_month'], amt_month['avg_purchase'],
           color='red', edgecolor='red', s=100, label='Средний чек', zorder=3)

max_check = amt_month['avg_purchase'].max()
am.set_ylim(bottom=0, top=max_check * 2)

am.set_title('Динамика среднего чека по месяцам')
am.set_xlabel('Номер месяца')
am.set_ylabel('Средний чек в рублях')
am.set_xticks(range(5, 13))
am.grid(True, linestyle='--', alpha=0.6)

am.legend(loc='upper left')

plt.show()
```
